```python
import pandas as pd
import os

# Llista de fitxers d'entrada
fitxers = ["presidents-at.csv",
"presidents-be.csv",
"presidents-bg.csv",
"presidents-cy.csv",
"presidents-cz.csv",
"presidents-de.csv",
"presidents-dk.csv",
"presidents-ee.csv",
"presidents-es.csv",
"presidents-fi.csv",
"presidents-fr.csv",
"presidents-gb.csv",
"presidents-gr.csv",
"presidents-hr.csv",
"presidents-hu.csv",
"presidents-ie.csv",
"presidents-it.csv",
"presidents-lt.csv",
"presidents-lu.csv",
"presidents-lv.csv",
"presidents-mt.csv",
"presidents-nl.csv",
"presidents-pl.csv",
"presidents-pt.csv",
"presidents-ro.csv",
"presidents-se.csv",
"presidents-si.csv",
"presidents-sk.csv",
"presidents-us.csv"
]

# Llista per acumular DataFrames
dfs = []

for fitxer in fitxers:
    # Llegeix el fitxer
    df = pd.read_csv(fitxer)
    
    # Afegeix codi ISO
    codi_iso = os.path.basename(fitxer).split("-")[1].split(".")[0]
    df.insert(0, "pais_iso", codi_iso)
    
    # Canvia el nom de columnes per asiminal
    df.rename(columns={"pais_iso":"iso_code",
                       "Nom": "nom",
                       "Nomenament": "nomenament",
                       "Partit": "posicio_politica"}, inplace=True)
    
    dfs.append(df)

# Combina tots els DataFrames
df_total = pd.concat(dfs, ignore_index=True)

# Mostrar els valors únics de la columna
valors_distints = df_total['posicio_politica'].unique()
print(valors_distints)

# Unificar valors de la columna
df_total['posicio_politica'] = df_total['posicio_politica'].str.lower()  # Passar tot a minúscules

# Substituir les variacions de les posicions per valors unificats
df_total['posicio_politica'] = df_total['posicio_politica'].replace({
    'eesquerra': 'esquerra',
    'independente': 'centre',
    'independent': 'centre',
    'derecha': 'dreta',
    'centre': 'centre',
    'cento': 'centre',
    'cente': 'centre',
})

# Mostrar els valors únics de la columna
valors_distints = df_total['posicio_politica'].unique()
print(valors_distints)
df_total.tail()
```

```python

# Discretitza la columna 'posicio_politica'
mapping = {'esquerra': 0, 'dreta': 1, 'centre': 2}
df_total['label'] = df_total['posicio_politica'].map(mapping)

# Desa el DataFrame combinat en un fitxer CSV
output_path = "presidents_combined_final.csv"
df_total.to_csv(output_path, index=False)

print(f"Fitxer combinat total registres: {df_total.shape}")
```
